## Implementation

Import necessary libraries and raw data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
# load 'iris' dataset
dataset = pd.read_csv('players_subset.csv')
dataset.info()
```

```python
dataset.head()
```

```python
features = ['Acceleration', 'Agility', 'Balance', 'Ball control', 'Composure', 'Crossing', 'Curve', 'Dribbling', 'Finishing', 'Free kick accuracy']
dataset[features].info()
dataset[features].head()
```

```python
X = dataset[features]
```

```python
kmeansmodel = KMeans(n_clusters=3)
kmeansmodel.fit(X)
centroids = kmeansmodel.cluster_centers_
centroids
```

```python
dataset['cluster'] = kmeansmodel.labels_
pd.DataFrame({'Name': dataset['Name'], 'Cluster': kmeansmodel.labels_})
```

```python
dataset_useful = dataset[['Name']+features+['cluster']]
dataset_useful.head()
```

```python
means = np.zeros((3,10))
for i in range(3):
    means[i] = dataset_useful[dataset_useful['cluster'] == i][features].mean()
    
```

```python
N=10
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ind = np.arange(N)
width = 0.3
# ax.bar(ind, men_means, width, bottom=0*cm, yerr=men_std, label='Men')
ax.bar(ind-width,means[0],width,label='Cluster 0')
ax.bar(ind,means[1],width,label='Cluster 1')
ax.bar(ind+width,means[2],width,label='Cluster 2')

ax.legend()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)

ax.set_xticks(ind)
ax.set_xticklabels(features)
ax.set_axisbelow(True)

plt.plot()
```

## Conclusion

From the result of K-Means clustering, we can see that players from cluster 0 performs worst in all the skills.Players in cluster 1 performs best in all skills.  Players in cluster 2 performs at a medium level in all skills.
